```python
gold_df=spark.table("gold_fraud_transactions")
gold_df.createOrReplaceTempView("gold_fraud_transactions")

```

Query 1: Total number of transactions

```python
%sql
select count(*) as gold_transactions from gold_fraud_transactions;
```

Query 2: Total suspicious transactions

```python
%sql
select count(*) as suspicious_transactions from gold_fraud_transactions where suspicious_flag='YES';
```

Query 3: High Value transactions

```python
%sql

SELECT txn_id,customer_name,amount FROM gold_fraud_transactions WHERE high_value_flag='YES' ORDER BY amount DESC;
```

```python
%sql

SELECT txn_id,customer_name,amount FROM gold_fraud_transactions WHERE international_flag='YES';
```

```python
%sql
SELECT account_id, customer_name, fraud_type FROM gold_fraud_transactions WHERE known_fraud_flag = 'YES';
```

```python
%sql
select customer_name,sum(amount) as total_amount from gold_fraud_transactions group by customer_name order by total_amount desc limit 10;
```

```python
%sql
SELECT branch,SUM(amount) AS total_amount FROM gold_fraud_transactions GROUP BY branch ORDER BY total_amount DESC;
```

```python
%sql
SELECT branch,COUNT(*) AS suspicious_transactions FROM gold_fraud_transactions WHERE suspicious_flag = 'YES' GROUP BY branch
ORDER BY suspicious_transactions DESC;
```

```python
%sql
SELECT txn_date,COUNT(*) AS total_transactions,SUM(amount) AS total_amount FROM gold_fraud_transactions GROUP BY txn_date
ORDER BY txn_date;
```

```python
%sql
SELECT fraud_type,COUNT(*) AS total_accounts FROM gold_fraud_transactions WHERE fraud_type IS NOT NULL GROUP BY fraud_type ORDER BY total_accounts DESC;
```
