```python
transactions_df=(spark.read.option("header","true").option("inferschema","true").csv("/Volumes/workspace/default/myfiles/transactions.csv"))
display(transactions_df)
```

```python
transactions_df.show(5)
transactions_df.printSchema()
transactions_df.count()
```

```python
fraud_df=(spark.read.option("header","true").option("inferschema","true").csv("/Volumes/workspace/default/myfiles/known_fraud_accounts.csv"))
display(fraud_df)
fraud_df.show(5)
fraud_df.printSchema()
fraud_df.count()
```

```python
accounts_df.write.format("delta").mode("overwrite").saveAsTable("bronze_accounts")
```

```python
transactions_df.write.format("delta").mode("overwrite").saveAsTable("bronze_transactions")
```

```python
fraud_df.write.format("delta").mode("overwrite").saveAsTable("bronze_frauds")
```
